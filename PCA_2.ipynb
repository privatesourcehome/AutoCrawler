```python
import pandas as pd
url = "C:/1234/x_train_money.csv"
df = pd.read_csv(url, names=['F1','F2','F3','F4','F5','D1','D2','D3','D4','D5','D6','D7','D8','D9','D10','M1'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler  # 표준화 패키지 라이브러리 
y = df['M1'].values # 종속변인 추출
x = df.drop(['M1'], axis=1).values # 독립변인들의 value값만 추출

features = ['F1','F2','F3','F4','F5','D1','D2','D3','D4','D5','D6','D7','D8','D9','D10']
pd.DataFrame(x, columns=features).head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=6) # 주성분을 몇개로 할지 결정
printcipalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data=printcipalComponents, columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6'])
# 주성분으로 이루어진 데이터 프레임 구성
```

```python
url = "C:/1234/x_train_money_after_pca.csv"
train = pd.read_csv(url, names=['F1','F2','F3','D1','D4','D6','M1'])
train.head()
```

```python
import numpy as np
y_train=np.array(train['M1'])
x_train=train.drop(['M1'], axis=1)
x_train=np.array(x_train)
```

```python
from sklearn.model_selection import KFold
```

```python
kfold = KFold(n_splits=5, shuffle = True, random_state=1234)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score, cross_validate

svm_model = SVR(kernel='rbf')
svm_model.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(svm_model , x_train , y_train ,cv=kfold)
print(scores)
```
